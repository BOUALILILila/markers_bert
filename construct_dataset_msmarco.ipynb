```python
import pandas as pd

data='/Data'

dfq = pd.read_csv(f'{data}/queries.train.tsv', delimiter='\t', header=None, names=['qid', 'query'])
print(dfq.shape)
dfq.sample(10)
```

```python
dfc = pd.read_csv(f'{data}/collection.tsv', delimiter='\t', header=None, names=['pid', 'passage'])
print(dfc.shape)
dfc.sample(10)
```

```python
dftrain = pd.read_csv(f'{data}/triples.train.small.tsv', delimiter='\t', header=None, names=['q_txt', 'pp_txt', 'np_txt'])
print(dftrain.shape)
dftrain.sample(10)
```

# Add label column from qrels (qrels file contains only relevant query-passage pairs label=1)

```python
dffinal1 = dftrain
dffinal1 = dffinal1[['q_txt','pp_txt']]
dffinal1.columns = ['q_txt','p_txt']
dffinal1['label'] = [1]*len(dffinal1)
dffinal2 = dftrain
dffinal2 = dffinal2[['q_txt','np_txt']]
dffinal2.columns = ['q_txt','p_txt']
dffinal2['label'] = [0]*len(dffinal2)
```

```python
dffinal = pd.concat([dffinal1,dffinal2])
dffinal 
```

```python
c=(dffinal.label==0)
neg=dffinal[c]
pos=dffinal[dffinal.label==1]
len(pos), len(neg)
```

# Discard redundunt pairs

```python
dff=dffinal.drop_duplicates(subset=['q_txt','p_txt'])
dff.shape
```

```python
dffinal.to_csv(f'{data}/triples.unique.train.small_full.pkl')
```
